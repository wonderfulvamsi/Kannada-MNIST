Data preprocessing :
===

```python
ytrain=np.array(train['label'])
```

```python
xtrainpics=train.drop(['label'], axis=1)
```

```python
xtrainpics
```

```python
xtrainpics=np.array(xtrainpics)
```

```python
xtrainpics=xtrainpics.reshape(60000,28,28,1)
```

```python
xvalpics=val.drop(['label'], axis=1)
```

```python
yval=val['label']
```

```python
xvalpics=np.array(xvalpics)
xvalpics.shape
```

```python
xvalpics=xvalpics.reshape(10240,28,28,1)
```

```python
yval=np.array(yval)
yval
```

```python
testpics=test.drop(['id'], axis=1)
```

Building the model :
===

```python
from keras.models import Sequential
from keras.layers import Dense,MaxPooling2D,BatchNormalization,Dropout,Flatten
from keras.layers import Convolution2D as Conv2D
```

```python
model = Sequential()
model.add(Conv2D(132, kernel_size = (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(BatchNormalization())
model.add(Conv2D(164, kernel_size=(3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(BatchNormalization())
model.add(Conv2D(164, kernel_size=(3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(BatchNormalization())
model.add(Conv2D(196, kernel_size=(3,3), activation='relu',padding='same'))
model.add(MaxPooling2D(pool_size=(2,2),padding='same'))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(200, activation='relu'))
model.add(Dropout(0.3))
# Output Layer
model.add(Dense(10,activation='softmax'))

model.summary()

# Compile the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=["accuracy"])
```

Training the model :
===

```python
model.fit(xtrainpics,ytrain,epochs=10,batch_size=2000,validation_data=(xvalpics,yval))
```

Saving the model :
===

```python
from sklearn.externals import joblib
joblib.dump(model,r'C:\Users\vamsi\Desktop\Kaggle2')
```

Making predictions :
===

```python
testpics=np.array(testpics)
testpics=testpics.reshape(5000,28,28,1)
sol=model.predict(testpics)
```

```python
solutions=[]
for i in range(len(sol)):
    solutions.append(sol[i].argmax())
```

```python
solutions=pd.DataFrame(solutions)
```
